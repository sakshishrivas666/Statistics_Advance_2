# 1. What is hypothesis testing in statistics ?

  - Hypothesis testing is a statistical method used to determine if there is enough evidence in a sample of data to support a certain belief or claim about a population. It involves setting up two competing statements about a population parameter: a null hypothesis (the status quo or no effect) and an alternative hypothesis (what you are trying to prove). You then collect and analyze data to calculate a test statistic, which is used to assess the strength of evidence against the null hypothesis. Based on the test statistic and the chosen significance level, you make a decision to either reject the null hypothesis or fail to reject it. This decision helps you draw conclusions about the population based on your sample data.

#2.  What is the null hypothesis, and how does it differ from the alternative hypothesis ?

  - The null hypothesis is the initial statement or claim made about a population that we are trying to disprove. It represents the status quo or the absence of an effect. The alternative hypothesis, on the other hand, is the statement we are trying to prove. It is accepted if we find sufficient evidence to reject the null hypothesis. In essence, the null hypothesis proposes no relationship or no difference, while the alternative hypothesis proposes a relationship or a difference.

#3.  What is the significance level in hypothesis testing, and why is it important ?

  - The significance level, denoted by alpha (α), is a crucial concept in hypothesis testing. It represents the probability of making a Type I error, which is the error of rejecting a true null hypothesis [1, 2].

# Threshold for decision-making:
The significance level sets the threshold for deciding whether the observed results are statistically significant enough to reject the null hypothesis. If the p-value (the probability of observing the data if the null hypothesis is true) is less than or equal to the significance level, we reject the null hypothesis.

#Controlling Type I error:
 By setting the significance level, researchers explicitly control the probability of incorrectly rejecting a true null hypothesis. A lower significance level reduces the risk of a Type I error.

#Interpreting results:
The significance level helps in interpreting the results of a hypothesis test. It provides a clear benchmark for determining whether the evidence from the sample is strong enough to support the alternative hypothesis.

#Standard practice:
Common significance levels like 0.05, 0.01, and 0.10 are widely used in various fields, allowing for consistency and comparison across studies [2].

#4.  What does a P-value represent in hypothesis testing ?

  - n hypothesis testing, a P-value represents the probability of observing the data, or data more extreme than the observed data, assuming the null hypothesis is true. It's a measure of the strength of evidence against the null hypothesis. A smaller P-value indicates stronger evidence against the null hypothesis. The P-value is compared to the chosen significance level (alpha) to make a decision about rejecting or failing to reject the null hypothesis. If the P-value is less than or equal to alpha, you reject the null hypothesis.



#5.  How do you interpret the P-value in hypothesis testing ?

# Interpreting the P-value in hypothesis testing

The P-value tells you the probability of observing your data (or data more extreme) if the null hypothesis were true. A small P-value suggests that your observed data is unlikely to occur if the null hypothesis is true, providing evidence against the null hypothesis. Conversely, a large P-value suggests that your observed data is reasonably likely to occur if the null hypothesis is true, so you don't have strong evidence against the null hypothesis.

You compare the P-value to your chosen significance level (alpha).

If the P-value is less than or equal to alpha, you reject the null hypothesis. This means your results are statistically significant, and you have evidence to support the alternative hypothesis.
If the P-value is greater than alpha, you fail to reject the null hypothesis. This means your results are not statistically significant, and you do not have enough evidence to support the alternative hypothesis.
It's important to remember that failing to reject the null hypothesis does not mean the null hypothesis is true, it simply means you don't have enough evidence to reject it based on your data.

#7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing  ?

  - n hypothesis testing, the difference between a one-tailed and a two-tailed test lies in the direction of the alternative hypothesis.

A one-tailed test is used when the alternative hypothesis specifies a direction. This means you are only interested in whether the population parameter is greater than or less than the null hypothesis value, but not both. For example, you might hypothesize that a new drug will increase a patient's recovery rate, so the alternative hypothesis is that the recovery rate is greater than the current rate. The rejection region for a one-tailed test is entirely on one side of the sampling distribution.

A two-tailed test is used when the alternative hypothesis does not specify a direction. You are interested in whether the population parameter is simply different from the null hypothesis value, regardless of whether it is greater or less. For example, you might hypothesize that a new teaching method has a different effect on student scores than the old method, so the alternative hypothesis is that the average score is not equal to the old average score. The rejection region for a two-tailed test is split between both sides of the sampling distribution.

The choice between a one-tailed and a two-tailed test depends on the specific question being asked and the nature of the alternative hypothesis. If you have a clear directional hypothesis, a one-tailed test is appropriate. If you are simply looking for any difference, a two-tailed test is the correct choice.



#8.  What is the Z-test, and when is it used in hypothesis testing ?

The Z-test is used when:

#The population variance is known:
 This is a key requirement for using a Z-test [1]. If the population variance is unknown and the sample size is small, a t-test is typically used instead.

#The data are approximately normally distributed:
The Z-test assumes that the data follows a normal distribution.

#The sample size is large (generally n > 30):
 Even if the population variance is unknown, for larger sample sizes, the Z-test can still be a reasonable approximation due to the Central Limit Theorem.

#9.  How do you calculate the Z-score, and what does it represent in hypothesis testing ?

  - To calculate the Z-score, you subtract the population mean from the sample mean, and then divide that result by the population standard deviation [1].

The Z-score represents how many standard deviations a data point is away from the mean of a distribution. In hypothesis testing, the Z-score (or Z-statistic) measures how many standard errors the sample mean is away from the hypothesized population mean under the null hypothesis. It helps determine how likely it is to observe the sample mean if the null hypothesis is true. A larger absolute Z-score indicates that the sample mean is further from the hypothesized population mean, providing stronger evidence against the null hypothesis.



#10.  What is the T-distribution, and when should it be used instead of the normal distribution ?

  - The T-distribution, also known as Student's t-distribution, is a probability distribution that is used in hypothesis testing when the sample size is small or the population standard deviation is unknown.

You should use the T-distribution instead of the normal distribution when:

#The population standard deviation is unknown:
The normal distribution relies on knowing the population standard deviation to calculate Z-scores. When you don't have this information and are working with a sample, the T-distribution is a better fit.

#The sample size is small:
For smaller sample sizes (generally less than 30), the T-distribution provides a more accurate representation of the sampling distribution of the mean compared to the normal distribution. This is because the T-distribution accounts for the increased variability that comes with smaller samples.

As the sample size increases, the T-distribution becomes more similar to the normal distribution. In fact, for very large sample sizes, the T-distribution is practically identical to the normal distribution.

#11.  What is the difference between a Z-test and a T-test ?

  - The Z-test and the T-test are both statistical tests used in hypothesis testing, but they differ in when they are appropriate to use.

The main difference lies in the information you have about the population standard deviation. A Z-test is used when you know the population standard deviation. In contrast, a T-test is used when you don't know the population standard deviation and instead have to estimate it from your sample data.

Another factor that influences the choice between a Z-test and a T-test is the sample size. When the sample size is large, even if you don't know the population standard deviation, the T-test results will be very similar to the Z-test. However, for small sample sizes, the T-distribution, which is used for the T-test, is wider and accounts for the increased uncertainty due to the smaller sample size. This makes the T-test more appropriate for smaller samples when the population standard deviation is unknown.

In summary, use a Z-test when you know the population standard deviation. Use a T-test when the population standard deviation is unknown and you have to rely on the sample standard deviation, especially when dealing with small sample sizes.

#12.  What is the T-test, and how is it used in hypothesis testing ?


  - The T-test is a statistical test used in hypothesis testing. It helps determine if there is a significant difference between the means of two groups, or between a sample mean and a known population mean, when the population standard deviation is unknown and the sample size is relatively small.

Here's how it's used in hypothesis testing:

#Formulate Hypotheses:
You start by stating a null hypothesis (H₀) and an alternative hypothesis (H₁). The null hypothesis typically states that there is no difference between the means. The alternative hypothesis states that there is a difference.

#Choose a Significance Level:
You select a significance level (alpha, α), which is the probability of rejecting the null hypothesis when it is actually true (Type I error). Common values are 0.05 or 0.01.

#Calculate the T-statistic:
Based on your sample data, you calculate a T-statistic. This statistic measures the difference between the sample means (or sample mean and population mean) relative to the variability within the samples.

#Determine Degrees of Freedom:
The T-test uses degrees of freedom, which are related to the sample size(s). This value is used to find the critical T-value or calculate the P-value.

#Find the P-value or Critical Value:
Using the T-statistic and degrees of freedom, you can either find the P-value (the probability of observing your data if the null hypothesis is true) or compare the T-statistic to a critical T-value from the T-distribution table.

#13.  What is the relationship between Z-test and T-test in hypothesis testing ?

  - The Z-test and the T-test are both statistical tools used for hypothesis testing, particularly for comparing means. Their relationship lies in the fact that the T-test is essentially an adaptation of the Z-test for situations where we have less information, specifically when the population standard deviation is unknown.

When we know the population standard deviation, the Z-test is the appropriate choice because the Z-distribution accurately reflects the sampling distribution of the mean. However, in many real-world scenarios, we don't have this information. Instead, we have to estimate the population standard deviation from our sample data.

This is where the T-test comes in. The T-test uses the sample standard deviation to estimate the population standard deviation. Because the sample standard deviation can vary from sample to sample, especially with small sample sizes, the T-distribution is used instead of the Z-distribution. The T-distribution is wider than the Z-distribution, reflecting the increased uncertainty due to using the sample standard deviation.

As the sample size increases, the sample standard deviation becomes a more reliable estimate of the population standard deviation. Consequently, the T-distribution approaches the Z-distribution. For large sample sizes, the results of a T-test and a Z-test will be very similar.

In summary, the T-test is a more robust test for situations with unknown population standard deviation, particularly for smaller sample sizes. It accounts for the additional variability introduced by estimating the population standard deviation from the sample. The Z-test is a special case used when the population standard deviation is known.

#14.  What is a confidence interval, and how is it used to interpret statistical results ?

  - A confidence interval is a range of values that is likely to contain the true population parameter with a certain level of confidence. It provides a measure of the uncertainty associated with a sample statistic.

Here's how it's used to interpret statistical results:

#Estimating population parameters:
Confidence intervals are used to estimate the range within which a population parameter (like the mean, proportion, or difference between means) is likely to fall. For example, a 95% confidence interval for the mean IQ score of a population might be [96.656, 106.422]. This means we are 95% confident that the true average IQ score for all students at that school is between 96.656 and 106.422 [2].

#Assessing precision:
A narrower confidence interval indicates a more precise estimate of the population parameter, while a wider interval suggests greater uncertainty. The width of the interval is influenced by the sample size and the variability of the data.

#Hypothesis testing:
Confidence intervals can be used to perform hypothesis tests. If a hypothesized value for the population parameter falls outside the confidence interval, it provides evidence to reject the null hypothesis.

#Making informed decisions:
In various fields, such as finance, confidence intervals help assess the reliability of estimates and inform decision-making [1]. For instance, in finance, they can be used to assess the reliability of estimated returns or risk measures

#15.  What is the margin of error, and how does it affect the confidence interval ?

  
  - The margin of error is the range of values above and below the sample statistic in a confidence interval. It quantifies the uncertainty in an estimate.

The margin of error directly affects the width of the confidence interval. A larger margin of error results in a wider confidence interval, indicating less precision in the estimate of the population parameter. Conversely, a smaller margin of error leads to a narrower confidence interval, suggesting a more precise estimate.

The margin of error is calculated based on the standard deviation or standard error of the sample statistic and the critical value from the appropriate distribution (like the Z-distribution or T-distribution) corresponding to the desired confidence level. Factors like sample size and variability in the data influence the margin of error. Larger sample sizes and lower variability tend to decrease the margin of error, resulting in narrower confidence intervals and more precise estimates.

#16.  How is Bayes' Theorem used in statistics, and what is its significance ?

  - Bayes' Theorem is used in statistics to update the probability of a hypothesis based on new evidence. It provides a way to revise existing beliefs (prior probabilities) in light of observed data to obtain updated beliefs (posterior probabilities). This approach is fundamental to Bayesian inference, where the theorem is used to determine the probability of a model or hypothesis given the data observed.

Its significance lies in its ability to incorporate prior knowledge or beliefs into statistical analysis and to provide a coherent framework for updating those beliefs as more data becomes available. This is particularly useful in situations where data is limited or where there are strong prior expectations. Bayes' Theorem allows for a more nuanced understanding of uncertainty by providing a probability distribution over possible hypotheses rather than just a simple rejection or acceptance of a null hypothesis, as in traditional frequentist methods. It is widely used in various fields, including machine learning, medical diagnosis, and finance.

#18.  What is the Chi-square goodness of fit test, and how is it applied ?

  - The Chi-square goodness of fit test is a statistical test used to determine if an observed frequency distribution of a single categorical variable differs significantly from an expected frequency distribution.

Here's how it is applied:

First, you state the null hypothesis (H₀) and the alternative hypothesis (H₁). The null hypothesis states that there is no significant difference between the observed and expected frequencies; in other words, the data fits the expected distribution. The alternative hypothesis states that there is a significant difference, and the data does not fit the expected distribution.

Next, you collect observed frequencies for each category. You also determine the expected frequencies for each category based on your null hypothesis or a theoretical distribution.

Then, you calculate the Chi-square test statistic. This involves summing the squared differences between the observed and expected frequencies, divided by the expected frequencies for each category.

After calculating the test statistic, you determine the degrees of freedom. For the goodness of fit test, the degrees of freedom are the number of categories minus one.

Finally, you compare the calculated Chi-square test statistic to a critical value from the Chi-square distribution table or find the P-value associated with the test statistic and degrees of freedom. If the test statistic is greater than the critical value, or if the P-value is less than your chosen significance level (alpha), you reject the null hypothesis. This indicates that there is a significant difference between the observed and expected frequencies, and the data does not fit the expected distribution. If you fail to reject the null hypothesis, it means there is not enough evidence to conclude that the observed frequencies differ significantly from the expected frequencies.

```python
#Q.2  Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range for the discrete uniform distribution (e.g., integers from 1 to 6)
low_val = 1
high_val = 7  # The upper bound is exclusive in scipy's randint

# Create a discrete uniform distribution object
dist = randint(low_val, high_val)

# Generate some random samples from the distribution
random_samples = dist.rvs(size=1000)

# Calculate the probability mass function (PMF)
# For a discrete uniform distribution, each value in the range has an equal probability.
# We can calculate this probability as 1 / (high_val - low_val).
possible_values = np.arange(low_val, high_val)
probabilities = dist.pmf(possible_values)

# Plot the PMF
plt.figure(figsize=(8, 6))
plt.bar(possible_values, probabilities, color='skyblue')
plt.xlabel("Possible Values")
plt.ylabel("Probability")
plt.title("Probability Mass Function (PMF) of a Discrete Uniform Distribution")
plt.xticks(possible_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Q.3  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pdf(p, x):
  """
  Calculates the Probability Distribution Function (PDF) of a Bernoulli distribution.

  Args:
    p: The probability of success (between 0 and 1).
    x: The value for which to calculate the PDF (either 0 for failure or 1 for success).

  Returns:
    The probability of the given value x occurring in a Bernoulli distribution
    with probability of success p. Returns 0 if x is not 0 or 1.
  """
  if x == 1:
    return p
  elif x == 0:
    return 1 - p
  else:
    return 0

# Example usage:
p_success = 0.7  # Probability of success
x_success = 1    # Value for success
x_failure = 0    # Value for failure

pdf_success = bernoulli_pdf(p_success, x_success)
pdf_failure = bernoulli_pdf(p_success, x_failure)

print(f"PDF at x=1 (success) with p={p_success}: {pdf_success}")
print(f"PDF at x=0 (failure) with p={p_success}: {pdf_failure}")

# Example with an invalid value for x
x_invalid = 2
pdf_invalid = bernoulli_pdf(p_success, x_invalid)
print(f"PDF at x={x_invalid} with p={p_success}: {pdf_invalid}")
```

```python
#Q.6 Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.

import numpy as np
import matplotlib.pyplot as plt

# Parameters of the discrete uniform distribution
a = 1  # lower bound
b = 10 # upper bound

# Create the range of values
x = np.arange(a, b + 1)

# Probability Mass Function (PMF) for discrete uniform distribution
pmf = np.ones_like(x) / (b - a + 1)

# Cumulative Distribution Function (CDF)
cdf = np.cumsum(pmf)

# Plotting the CDF
plt.figure(figsize=(8, 5))
plt.step(x, cdf, where='post', label='CDF', color='blue')
plt.scatter(x, cdf, color='red')
plt.title('CDF of Discrete Uniform Distribution')
plt.xlabel('x')
plt.ylabel('CDF')
plt.xticks(x)
plt.grid(True)
plt.legend()
plt.show()

# Print values for clarity
print("x-values:", x)
print("CDF:", np.round(cdf, 2))
```

```python
# Q.16  Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats

# Sample dataset
data = [12, 15, 14, 10, 13, 12, 15, 14, 13, 16]

# Calculate sample mean and standard error
mean = np.mean(data)
std_err = stats.sem(data)  # Standard error of the mean

# Define confidence level
confidence = 0.95

# Calculate the margin of error using t-distribution
degrees_freedom = len(data) - 1
t_critical = stats.t.ppf((1 + confidence) / 2, degrees_freedom)
margin_of_error = t_critical * std_err

# Calculate confidence interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"Sample mean: {mean:.2f}")
print(f"{int(confidence*100)}% confidence interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
# Q.3  Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Calculate Z score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate p-value for two-tailed test
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Decision
    reject_null = p_value < alpha

    return {
        'sample_mean': sample_mean,
        'z_score': z_score,
        'p_value': p_value,
        'reject_null': reject_null
    }

# Example Usage
sample_data = [12, 15, 14, 10, 13, 14, 16, 11, 13, 12]
population_mean = 13
population_std = 2  # Known population std dev

result = one_sample_z_test(sample_data, population_mean, population_std)
print(result)
```

```python
# Q.5  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(alpha=0.05, mu0=0, mu1=1, sigma=1, two_tailed=False):
    """
    Visualize Type I and Type II errors for a z-test.

    Parameters:
    - alpha: significance level (Type I error rate)
    - mu0: mean under the null hypothesis
    - mu1: mean under the alternative hypothesis
    - sigma: standard deviation
    - two_tailed: whether the test is two-tailed
    """
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)

    # Null and alternative distributions
    null_dist = norm(mu0, sigma)
    alt_dist = norm(mu1, sigma)

    # Determine critical value(s)
    if two_tailed:
        z_alpha = norm.ppf(1 - alpha / 2)
        critical_low = mu0 - z_alpha * sigma
        critical_high = mu0 + z_alpha * sigma
    else:
        z_alpha = norm.ppf(1 - alpha)
        critical_low = None
        critical_high = mu0 + z_alpha * sigma

    # Plot distributions
    plt.figure(figsize=(10, 6))
    plt.plot(x, null_dist.pdf(x), label='Null Hypothesis (H₀)', color='blue')
    plt.plot(x, alt_dist.pdf(x), label='Alternative Hypothesis (H₁)', color='red')

    # Shade Type I error region(s)
    if two_tailed:
        x_alpha_low = x[x < critical_low]
        x_alpha_high = x[x > critical_high]
        plt.fill_between(x_alpha_low, 0, null_dist.pdf(x_alpha_low), color='blue', alpha=0.3, label='Type I Error (α)')
        plt.fill_between(x_alpha_high, 0, null_dist.pdf(x_alpha_high), color='blue', alpha=0.3)
    else:
        x_alpha = x[x > critical_high]
        plt.fill_between(x_alpha, 0, null_dist.pdf(x_alpha), color='blue', alpha=0.3, label='Type I Error (α)')

    # Shade Type II error region
    if two_tailed:
        x_beta = x[(x >= critical_low) & (x <= critical_high)]
    else:
        x_beta = x[x < critical_high]
    plt.fill_between(x_beta, 0, alt_dist.pdf(x_beta), color='red', alpha=0.3, label='Type II Error (β)')

    # Calculate β (Type II error rate)
    if two_tailed:
        beta = alt_dist.cdf(critical_high) - alt_dist.cdf(critical_low)
    else:
        beta = alt_dist.cdf(critical_high)

    power = 1 - beta

    # Plot formatting
    plt.title(f'Type I and Type II Errors\nα = {alpha:.2f}, β = {beta:.2f}, Power = {power:.2f}')
    plt.xlabel('Test Statistic')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    return beta, power

visualize_type1_type2_errors(alpha=0.05, mu0=0, mu1=1, sigma=1, two_tailed=False)

```

```python
# Q.9  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = sum(sample) / n
    std_err = stats.sem(sample)
    margin = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)
    lower_bound = mean - margin
    upper_bound = mean + margin
    return lower_bound, upper_bound

# This is your sample data — you can replace these numbers with your own data
sample_data = [12, 15, 14, 16, 13, 15, 14, 15, 16, 14]

# Calculate the confidence interval
ci = confidence_interval(sample_data)

# Print it out so you see the answer
print(f"95% Confidence interval: {ci}")
```

```python
# Q.10 Write a Python program to calculate the margin of error for a given confidence level using sample data.

import math
import scipy.stats as stats

def margin_of_error(data, confidence_level):
    n = len(data)
    mean = sum(data) / n
    std_dev = math.sqrt(sum((x - mean) ** 2 for x in data) / (n - 1))  # sample std dev

    # t critical value for two-tailed test
    t_crit = stats.t.ppf((1 + confidence_level) / 2, df=n-1)

    # margin of error
    margin_error = t_crit * (std_dev / math.sqrt(n))
    return margin_error

# Example usage:
sample_data = [12, 15, 14, 10, 13, 16, 12, 14]
confidence = 0.95

me = margin_of_error(sample_data, confidence)
print(f"Margin of Error at {confidence*100}% confidence level: {me:.4f}")
```

```python
# Q.11  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(prior, likelihood, marginal_likelihood):
    return (likelihood * prior) / marginal_likelihood

# Given probabilities
P_D = 0.01            # Prior: Probability of having disease
P_not_D = 1 - P_D     # Probability of not having disease

P_pos_given_D = 0.99  # Likelihood: Probability test positive given disease
P_pos_given_not_D = 0.05  # Probability test positive given no disease

# Marginal likelihood: total probability of testing positive
P_pos = (P_pos_given_D * P_D) + (P_pos_given_not_D * P_not_D)

# Posterior: Probability of disease given positive test
P_D_given_pos = bayes_theorem(P_D, P_pos_given_D, P_pos)

print(f"Probability of disease given positive test: {P_D_given_pos:.4f}")
```

```python
# Q.12 Perform a Chi-square test for independence between two categorical variables in Python.

import pandas as pd
from scipy.stats import chi2_contingency

# Sample data: contingency table
# Let's say we are checking the relationship between Gender and Preference
data = {
    'Preference_A': [30, 20],  # e.g. Male: 30, Female: 20
    'Preference_B': [10, 40]   # e.g. Male: 10, Female: 40
}
# Rows = Gender, Columns = Preference
# Index: Male, Female
df = pd.DataFrame(data, index=['Male', 'Female'])

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(df)

# Output results
print("Chi-square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-value:", p)
print("Expected Frequencies:\n", expected)

# Interpretation
alpha = 0.05
if p < alpha:
    print("Conclusion: Reject the null hypothesis. There is a significant association between the variables.")
else:
    print("Conclusion: Fail to reject the null hypothesis. No significant association between the variables.")
```

```python
# Q.13  Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    # Convert to NumPy array in case input is a list
    observed = np.array(observed)

    if observed.ndim == 1:
        # One-dimensional case (goodness-of-fit)
        total = observed.sum()
        expected = np.full_like(observed, total / len(observed))
    elif observed.ndim == 2:
        # Two-dimensional case (test of independence)
        row_totals = observed.sum(axis=1, keepdims=True)
        col_totals = observed.sum(axis=0, keepdims=True)
        grand_total = observed.sum()
        expected = (row_totals @ col_totals) / grand_total
    else:
        raise ValueError("Observed data must be 1D or 2D.")

    return expected

# Example usage:
observed_data = [
    [10, 20, 30],
    [6,  9,  15]
]

expected = calculate_expected_frequencies(observed_data)

print("Observed Frequencies:")
print(np.array(observed_data))
print("\nExpected Frequencies:")
print(np.round(expected, 2))
```

```python
# Q.14  Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

import scipy.stats as stats

# Observed frequencies
observed = [8, 9, 10, 11, 12, 10]

# Expected frequencies for a fair die
expected = [10, 10, 10, 10, 10, 10]

# Perform Chi-Square Goodness-of-Fit test
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")
```
